```python
import numpy as np
import pandas as pd
import time

cat_name_agent = ['Docked', 'Arrived', 'Counting','QC', 'Labeling', 'Received', 'Putaway', 'Picking', 'Packing',
                  'AWB', 'RTS', 'RT_picking', 'RT_putaway', 'Cyclecount']

productivity_varable = {
    'DL%':1,
    'DL % threshold':0.6,
    'Docked':75,
    'Arrived':125,
    'QC':4638,
    'Labeling':850,
    'Received':800,
    'Putaway':65,
    'Putaway_4floor':65,
    'Picking':114,
    'Packing':143,
    'Counting':1000,
    'AWB':720,
    'RTS':300,
    'RT_picking':726,
    'RT_putaway':726,
    'RT_picking_4floor':726,
    'RT_putaway_4floor':726,
    'Cyclecount':850,
    'Cyclecount_4floor':850
}

team_prod_dict = {
    'Picking':'出貨控場',
    'Packing':'出貨控場',
    'AWB':'出貨控場',
    'Arrived':'進貨控場',
    'Counting':'進貨控場',
    'QC':'進貨控場',
    'Labeling':'進貨控場',
    'Received':'進貨控場',
    'Docked':'進貨控場',
    # 'Print':'進貨控場',
    'RT_picking':'移庫控場',
    'RT_putaway':'移庫控場',
    'RT_picking_4floor':'移庫控場_四樓',
    'RT_putaway_4floor':'移庫控場_四樓',
    'RTS':np.nan,
    'Putaway':'上架控場',
    'Putaway_4floor':'上架控場_四樓',
    'Cyclecount':'盤點控場',
    'Cyclecount_4floor':'盤點控場_四樓'
}

whole_df = pd.read_csv('whole_df.csv', index_col=0, parse_dates=['create_time'])
punch_df = pd.read_csv('punch_df.csv', parse_dates=['date', 'created_time', 'end_time'])
```

```python
whole_df['date'] = whole_df['create_time'].dt.date

work_time = punch_df.groupby(['date', 'ID', 'type'])['hour'].agg([np.sum])\
                    .rename(columns={'sum':'hour'})\
                    .reset_index()
work_time['merge_type'] = work_time['type'].str.replace('_4floor', '')

total_pcs = whole_df.groupby(['date', 'operator', 'type'])['total_pcs'].agg([np.sum])\
                    .rename(columns={'sum':'total_pcs'})\
                    .reset_index()

work_time['date'] = work_time['date'].astype('str')  # 轉為字串形式以利合併
total_pcs['date'] = total_pcs['date'].astype('str')  # 轉為字串形式以利合併

productivity = work_time.merge(total_pcs, left_on=['date', 'ID', 'merge_type'], right_on=['date', 'operator', 'type'], how='inner')\
                        .rename(columns={'type_x':'type'})

productivity = productivity[['date', 'ID', 'type', 'hour', 'operator', 'total_pcs']]
productivity['target'] = productivity['type'].apply(lambda type_:productivity_varable[type_])
productivity['team'] = productivity['type'].apply(lambda type_:team_prod_dict[type_])
productivity['meet_target'] = ((productivity['total_pcs'] / productivity['hour']) >= productivity['target']).astype('int')  # 每小時完成數量比目標多，1是有達標、0是沒達標

# 1. productivity_raw: 每天每個category的工作人數、達標人數、達標率(呈現在同一張表)
productivity_raw = productivity.groupby(['team', 'type', 'date'])['meet_target'].agg([np.sum, 'count']).reset_index()
productivity_raw.columns = ['function', 'type', 'create_date', 'arrive_thres', 'count']
productivity_raw['prod_day_ratio'] = productivity_raw['arrive_thres'] / productivity_raw['count']

# 2~4. 每天每個category的工作人數(productivity_ppl_category)、達標人數(productivity_meet_category)與達標率(productivity_mean_category)
productivity_ppl_category = pd.crosstab(productivity['date'], productivity['type'])
productivity_meet_category = pd.crosstab(productivity['date'], productivity['type'],
                                         values=productivity['meet_target'], aggfunc=np.sum).fillna(0)
productivity_mean_category = productivity_meet_category / productivity_ppl_category

# 5~7. 每天每個function的工作人數(productivity_ppl_function)、達標人數(productivity_meet_function)與達標率(productivity_mean_function)
productivity_ppl_function = pd.crosstab(productivity['date'], productivity['team'])
productivity_meet_function = pd.crosstab(productivity['date'], productivity['team'],
                                         values=productivity['meet_target'], aggfunc=np.sum).fillna(0)
productivity_mean_function = productivity_meet_function / productivity_ppl_function

# 8. productivity_team_function: 呈現每天每個function的管理人員達標率
productivity_function = productivity.groupby(['team', 'date'])['meet_target'].agg([np.sum, 'count']).reset_index()
productivity_function.columns = ['function', 'create_date', 'arrive_thres', 'count']
productivity_function['prod_day_ratio'] = productivity_function['arrive_thres'] / productivity_function['count']
productivity_function['create_date'] = productivity_function['create_date'].astype('str')

prod_day_ratio = punch_df[punch_df['function_name'] == 'MGMT'][['ID', 'name', 'date', 'function']].drop_duplicates()
prod_day_ratio['date'] = prod_day_ratio['date'].astype('str')
prod_day_ratio = prod_day_ratio.merge(productivity_function, left_on=['date', 'function'], right_on=['create_date', 'function'])
productivity_team_function = pd.crosstab([prod_day_ratio['ID'], prod_day_ratio['name'], prod_day_ratio['function']],
                                         prod_day_ratio['date'],
                                         values=prod_day_ratio['prod_day_ratio'], aggfunc='mean')
productivity_team_function['date_on_duty'] = productivity_team_function.count(axis=1)

# 9. productivity_tl : 計算每個管理人員一個月的績效平均
productivity_tl = prod_day_ratio.groupby(['ID', 'name'])['prod_day_ratio'].agg(['count', np.mean])\
                                .rename(columns={'count':'days_on_duty', 'mean':'TL_produtivity_score'})

# 10. team_df
team_df = punch_df[punch_df['function_name'] == 'MGMT']
team_df = team_df[['ID', 'name', 'function', 'date', 'role', 'class', 'group', 'Unnamed: 7', 'function_name', 'function_role', 'min', 'hour']]
team_df = team_df.groupby(['ID', 'name', 'function', 'date', 'role', 'class', 'group', 'Unnamed: 7', 'function_name', 'function_role'])[['min', 'hour']].agg([np.sum])\
                 .reset_index()
team_df.columns = ['ID', 'name', 'function', 'date', 'role', 'class', 'group', 'Unnamed: 7', 'function_name', 'function_role', 'min', 'hour']
team_df['date'] = team_df['date'].astype('str')
team_df = team_df.merge(productivity_function, left_on=['function', 'date'], right_on=['function', 'create_date'], how='left')
```

```python
def get_prod_TL_score(productivity_varable, team_prod_dict, whole_df, punch_df, tl_output_path):
    whole_df['date'] = whole_df['create_time'].dt.date

    work_time = punch_df.groupby(['date', 'ID', 'type'])['hour'].agg([np.sum])\
                        .rename(columns={'sum':'hour'})\
                        .reset_index()
    work_time['merge_type'] = work_time['type'].str.replace('_4floor', '')

    total_pcs = whole_df.groupby(['date', 'operator', 'type'])['total_pcs'].agg([np.sum])\
                        .rename(columns={'sum':'total_pcs'})\
                        .reset_index()

    work_time['date'] = work_time['date'].astype('str')  # 轉為字串形式以利合併
    total_pcs['date'] = total_pcs['date'].astype('str')  # 轉為字串形式以利合併

    productivity = work_time.merge(total_pcs, left_on=['date', 'ID', 'merge_type'], right_on=['date', 'operator', 'type'], how='inner')\
                            .rename(columns={'type_x':'type'})

    productivity = productivity[['date', 'ID', 'type', 'hour', 'operator', 'total_pcs']]
    productivity['target'] = productivity['type'].apply(lambda type_:productivity_varable[type_])
    productivity['team'] = productivity['type'].apply(lambda type_:team_prod_dict[type_])
    productivity['meet_target'] = ((productivity['total_pcs'] / productivity['hour']) >= productivity['target']).astype('int')  # 每小時完成數量比目標多，1是有達標、0是沒達標

    # 1. productivity_raw: 每天每個category的工作人數、達標人數、達標率(呈現在同一張表)
    productivity_raw = productivity.groupby(['team', 'type', 'date'])['meet_target'].agg([np.sum, 'count']).reset_index()
    productivity_raw.columns = ['function', 'type', 'create_date', 'arrive_thres', 'count']
    productivity_raw['prod_day_ratio'] = productivity_raw['arrive_thres'] / productivity_raw['count']

    # 2~4. 每天每個category的工作人數(productivity_ppl_category)、達標人數(productivity_meet_category)與達標率(productivity_mean_category)
    productivity_ppl_category = pd.crosstab(productivity['date'], productivity['type'])
    productivity_meet_category = pd.crosstab(productivity['date'], productivity['type'],
                                             values=productivity['meet_target'], aggfunc=np.sum).fillna(0)
    productivity_mean_category = productivity_meet_category / productivity_ppl_category

    # 5~7. 每天每個function的工作人數(productivity_ppl_function)、達標人數(productivity_meet_function)與達標率(productivity_mean_function)
    productivity_ppl_function = pd.crosstab(productivity['date'], productivity['team'])
    productivity_meet_function = pd.crosstab(productivity['date'], productivity['team'],
                                             values=productivity['meet_target'], aggfunc=np.sum).fillna(0)
    productivity_mean_function = productivity_meet_function / productivity_ppl_function

    # 8. productivity_team_function: 呈現每天每個function的管理人員達標率
    productivity_function = productivity.groupby(['team', 'date'])['meet_target'].agg([np.sum, 'count']).reset_index()
    productivity_function.columns = ['function', 'create_date', 'arrive_thres', 'count']
    productivity_function['prod_day_ratio'] = productivity_function['arrive_thres'] / productivity_function['count']
    productivity_function['create_date'] = productivity_function['create_date'].astype('str')

    prod_day_ratio = punch_df[punch_df['function_name'] == 'MGMT'][['ID', 'name', 'date', 'function']].drop_duplicates()
    prod_day_ratio['date'] = prod_day_ratio['date'].astype('str')
    prod_day_ratio = prod_day_ratio.merge(productivity_function, left_on=['date', 'function'], right_on=['create_date', 'function'])
    productivity_team_function = pd.crosstab([prod_day_ratio['ID'], prod_day_ratio['name'], prod_day_ratio['function']],
                                             prod_day_ratio['date'],
                                             values=prod_day_ratio['prod_day_ratio'], aggfunc='mean')
    productivity_team_function['date_on_duty'] = productivity_team_function.count(axis=1)

    # 9. productivity_tl : 計算每個管理人員一個月的績效平均
    productivity_tl = prod_day_ratio.groupby(['ID', 'name'])['prod_day_ratio'].agg(['count', np.mean])\
                                    .rename(columns={'count':'days_on_duty', 'mean':'TL_produtivity_score'})

    # 10. team_df
    team_df = punch_df[punch_df['function_name'] == 'MGMT']
    team_df = team_df[['ID', 'name', 'function', 'date', 'role', 'class', 'group', 'Unnamed: 7', 'function_name', 'function_role', 'min', 'hour']]
    team_df = team_df.groupby(['ID', 'name', 'function', 'date', 'role', 'class', 'group', 'Unnamed: 7', 'function_name', 'function_role'])[['min', 'hour']].agg([np.sum])\
                     .reset_index()
    team_df.columns = ['ID', 'name', 'function', 'date', 'role', 'class', 'group', 'Unnamed: 7', 'function_name', 'function_role', 'min', 'hour']
    team_df['date'] = team_df['date'].astype('str')
    team_df = team_df.merge(productivity_function, left_on=['function', 'date'], right_on=['function', 'create_date'], how='left')

    productivity_ppl_category.reset_index(inplace=True)
    productivity_meet_category.reset_index(inplace=True)
    productivity_mean_category.reset_index(inplace=True)
    productivity_mean_function.reset_index(inplace=True)
    productivity_team_function.reset_index(inplace=True)
    productivity_tl.reset_index(inplace=True)

    with pd.ExcelWriter(tl_output_path) as writer:  
        team_df.to_excel(writer, sheet_name='team_df', index=False)
        productivity_tl.to_excel(writer, sheet_name='productivity_tl', index=False)
        productivity_raw.to_excel(writer, sheet_name='productivity_raw', index=False)
        productivity_ppl_category.to_excel(writer, sheet_name='productivity_ppl_category', index=False)
        productivity_meet_category.to_excel(writer, sheet_name='productivity_meet_category', index=False)
        productivity_mean_category.to_excel(writer, sheet_name='productivity_mean_category', index=False)
        productivity_mean_function.to_excel(writer, sheet_name='productivity_mean_function', index=False)
        productivity_team_function.to_excel(writer, sheet_name='productivity_team_function', index=False)

get_prod_TL_score(productivity_varable, team_prod_dict, whole_df, punch_df, 'TL_test.xlsx')
print('Checkpoint 7 productivity_TL SUCCEED')
```
