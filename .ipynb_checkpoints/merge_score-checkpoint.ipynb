```python
import time
import os
import numpy as np
import pandas as pd
import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
file_name = os.listdir("Output/March/productivity_agent_valid")
```

```python
merge_df = pd.DataFrame()

for file in file_name:
    date = file.replace("productivity_agent_", "").replace("_valid", "").replace(".xlsx", "")
    df = pd.read_excel(os.path.join("Output/March/productivity_agent_valid", file))
    df = df[["ID", "name", "class", "Productivity Score"]].rename(columns={"Productivity Score": date})
    if merge_df.empty:
        merge_df = df
    else:
        merge_df = pd.merge(merge_df, df, how="outer")
```

```python
index = merge_df[["ID", "name", "class"]]
score = merge_df.reindex(sorted(merge_df.columns[3:]), axis=1)
```

```python
output = pd.concat([index, score], axis=1).fillna(0)
output.to_excel("March_score.xlsx", index=False, encoding="utf_8")
```
